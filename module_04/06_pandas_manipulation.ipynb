```python
# Selecting rows - loc and iloc, loc uses labels, iloc uses integer positions


# iloc - by position 
df.iloc[0]  # first row
```

```python
# First 5 rows , first 3 columns 
df.iloc[:5, :3]
```

```python
# loc - by label

# First 5 rows, specific columns by name
df.loc[:4, ['Hours_Studied', 'Exam_Score', 'Gender']]
```

```python
# Filtering rows

# Students who scored above 70
high_scorers = df[df['Exam_Score'] > 70]
print(f"High Scorers: {len(high_scorers)} out of {len(df)}")
```

```python
df['Exam_Score'] > 70
```

```python
high_scorers.head(10)
```

```python
# Multiple conditions
# Students who scored above 70 and studied more than 25 hours
top_students = df[(df['Exam_Score'] > 70) & (df['Hours_Studied'] > 25)]
print(f"Top Students: {len(top_students)} out of {len(df)}")
```

```python
top_students.head()
```

```python
# isin() - Matching a List
# isin() allows you to check if values in a column are in a list of values
motivated = df[df['Motivation_Level'].isin(["High", "Medium"])]
print(f"Motivated Students: {len(motivated)} out of {len(df)}")
```

```python
motivated.head(10)
```

```python
# query() - The clean way
# query() - SQL - like filtering
result = df.query("Exam_Score > 70 and Hours_Studied > 25 and Gender == 'Female'")
print(f"Result: {len(result)} female students")
```

```python
result.head()
```

```python
# Sorting
# Sort by Exam_Score - highest first
df.sort_values('Exam_Score', ascending=False).head(10)
```

```python
# Sort by Multiple Columns 
df.sort_values(['Gender', 'Exam_Score'], ascending=[True, False]).head(2700)
```

```python
#  Adding, Renaming, and Dropping Columns

# Add a new column: Study Efficiency = Exam Score / Hours Studied
df['Study_Efficiency'] = df['Exam_Score'] / df['Hours_Studied']
df[['Exam_Score', 'Hours_Studied', 'Study_Efficiency']].head(10)
```

```python
df.shape
```

```python
# REnameing Columns
df = df.rename(columns={'Hours_Studied': 'Study_Hours', 'Exam_Score': 'Score'})
df.columns.tolist()
```

```python
# Drop columns
df = df.drop(columns=['Study_Efficiency'])
df.columns.tolist()
```

```python
df.shape
```

```python
# Renameing back
df = df.rename(columns={'Study_Hours': 'Hours_Studied', 'Score': 'Exam_Score'})
df.columns.tolist()
```
