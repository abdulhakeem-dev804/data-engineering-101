```python
import pandas as pd
```

# Series

## What is a series ?
- A series is a one-dimensional Array with lables . 

```python
s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

```python
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)
```

# DataFrame

## What is a DataFrame ?
- Two Dimensional Table - rows and columns. 

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
df
```
