```python
import numpy as np


# Create a NUmpy array from a python list
scores = np.array([85, 92, 78, 96, 88, 73, 91, 82, 89, 95])
print(scores)
print(type(scores))
```

```python
# Quick array creators
zeros = np.zeros(5)
ones = np.ones(5)
sequence = np.arange(0, 50, 10)
even_spread = np.linspace(0, 100, 5)
print(zeros)
print(ones)
print(sequence)
print(even_spread)
```

```python
# Array Properties
print("Shape:", scores.shape)
print("DType:", scores.dtype)
print("Dimensions:", scores.ndim)
print("Size:", scores.size)
```

```python
# Numppy Operations 

## Element - wise math 
print("Original Scores:", scores)
print("Add 5: ", scores + 5)
print("Multiply by 2: ", scores * 2)
print("Squared: ", scores ** 2)
```

```python
# Aggreation - Single value from array 
print(f"Mean: {scores.mean()}")
print(f"Sum: {scores.sum()}")
print(f"Max: {scores.max()}")
print(f"Min: {scores.min()}")
print(f"Std: {scores.std()}")
```

```python
# Boolean Filtering - just like Pandas !
print("Scores > 85:", scores[scores > 85])
print("Count > 85:", np.sum(scores > 85))
```

```python
# Numpy + Pandas Connection
import pandas as pd

df = pd.read_csv('data/StudentPerformanceFactors.csv')
```

```python
# Convert a pandas column to Numpy array
exam_scores = df['Exam_Score'].to_numpy()
print(f"Type :  {type(exam_scores)}")
print(f"Mean: {exam_scores.mean()}")
print(f"Above 80: {np.sum(exam_scores > 80)} students")
```

```python
# Matplotlib - Visualization Library

import matplotlib.pyplot as plt

# Simple Line Plot
study_hours = [5, 10, 15, 20, 25, 30, 35]
avg_scores =  [55, 62, 70, 75, 80, 85, 88]

plt.figure(figsize=(8, 5))
plt.plot(study_hours, avg_scores, marker='o', color='#2196F3', linewidth=2)
plt.title('Study Hours vs Average Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Average Exam Score')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Chars Types

# Histogram - Distibution 
# Histogram - see the distibution of exam scores
plt.figure(figsize=(8, 5))
plt.hist(df['Exam_Score'], bins=20, color='#4CAF50', edgecolor='white')
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Bar Chart - Comparison

# Bar Chart - average score by motivation level
avg_by_motivation = df.groupby('Motivation_Level')['Exam_Score'].mean() 

plt.figure(figsize=(8, 5))
plt.bar(avg_by_motivation.index, avg_by_motivation.values, color=['#FF5722','#FF9800', '#4CAF50'])
plt.title('Average Exam Score by Motivation Level')
plt.xlabel('Motivation Level')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
# Scatter plot - Relationship between two variables

# relationship between hours studied and exam score
plt.figure(figsize=(8, 5))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], alpha=0.3, color='#9C27B0', s=10)
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```
